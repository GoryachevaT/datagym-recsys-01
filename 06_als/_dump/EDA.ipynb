# Бейзлайн: глобальный топ

```python
#from tqdm.notebook import tqdm
from collections import defaultdict
```

```python
observations = defaultdict(int)

# почему круглые скобки внутри tqdm?
#for js in tqdm((json.loads(s) for s in open('./data/user_history.jsons'))):
for js in (json.loads(s) for s in open('./data/user_history.jsons')):
    for trans in js['transaction_history']:
        for product in trans["products"]:
            observations[product["product_id"]] += 1

_tmp = list(observations.keys())
# зачем минус в лямбде?
top_products = sorted(_tmp, key=lambda x: -observations[x])
```

# Проверяем качество

```python
from src.metrics import normalized_average_precision

scores = []
#for js in tqdm((json.loads(s) for s in open('./data/user_history.jsons'))):
for js in (json.loads(s) for s in open('./data/user_history.jsons')):
    recommended_items = top_products[:30]
    # какой физический смысл у этого действия?
    gt_items = js["target"][0]["product_ids"]
    ap = normalized_average_precision(gt_items, recommended_items)
    scores.append(ap)
print("score: {}".format(np.mean(scores)))
```

# Перенумеровываем данные для дальнейшей работы

```python
product_encoder = {}
# А какие есть альтернативы .itertuples()?
for i, row in enumerate(product_df.itertuples()):
    product_encoder[row.product_id] = i
```

```python
train_data = []
valid_data = []

#for js in tqdm((json.loads(s) for s in open('./data/user_history.jsons'))):
for js in (json.loads(s) for s in open('./data/user_history.jsons')):
    train_ids = []
    for transaction in js["transaction_history"]:
        train_ids.extend([product_encoder[x["product_id"]] for x in transaction["products"]])
    
    target_ids = [product_encoder[x] for x in js["target"][0]["product_ids"]]
    
    train_data.append(np.array(train_ids))
    valid_data.append(np.array(target_ids))
```

```python
np.savez_compressed("./tmp/data.npz", train_ids=train_data, valid_ids=valid_data)
```
